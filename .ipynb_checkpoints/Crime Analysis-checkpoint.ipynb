## Example Queries to Run for Analysis on Star Schema

```python
import pandas as pd
import mysql.connector
```

#### The City where the MOST Violent Crimes were recorded

```python
conn = mysql.connector.connect (
    host = "",
    user = "",
    passwd = "",
    database = ""
    )
cur = conn.cursor()

query = pd.read_sql_query("""SELECT DISTINCT cd.City, ft.ViolentCrime
            FROM fact_table as ft
            INNER JOIN city_dim as cd
            ON cd.City_id = ft.City_id
            WHERE ViolentCrime = (SELECT MAX(ViolentCrime)
                                   FROM fact_table
                                  )""",conn)
query.head()
```

#### Top 10 Cities where LEAST Violent Crimes were recorded

```python
query = pd.read_sql_query("""SELECT DISTINCT cd.City, ft.ViolentCrime
            FROM fact_table as ft
            INNER JOIN city_dim as cd
            ON cd.City_id = ft.City_id
            ORDER BY ViolentCrime ASC
            """,conn)
query.head(10)
```

#### Top 10 Cities for most Violent Crime offenses

```python
query = pd.read_sql_query("""SELECT DISTINCT cd.City, ft.Year_id, ft.ViolentCrime
                            FROM fact_table as ft
                            INNER JOIN city_dim as cd
                            ON cd.City_id = ft.City_id
                            ORDER BY ViolentCrime DESC""",conn)

query.head(10)
```

#### Top 10 Cities where most Arson offenses happen

```python
query = pd.read_sql_query("""SELECT DISTINCT cd.City, ft.Year_id, ft.Arson
                            FROM fact_table as ft
                            INNER JOIN city_dim as cd
                            ON cd.City_id = ft.City_id
                            ORDER BY Arson DESC""",conn)

query.head(10)
```

#### Top 10 States where Robbery offenses were recorded

```python
query = pd.read_sql_query("""SELECT DISTINCT cd.City, ft.Year_id, ft.Robbery
                            FROM fact_table as ft
                            INNER JOIN city_dim as cd
                            ON cd.City_id = ft.City_id
                            ORDER BY Robbery DESC""",conn)

query.head(10)
```
